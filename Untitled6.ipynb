```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# 1. LOAD DATASET
# Ensure the filename matches exactly what you uploaded
df = pd.read_csv('wine-quality-white-and-red.csv')

print("--- Wine Data Preview ---")
print(df.head())

# 2. PRE-PROCESSING
# We drop 'type' (text) and 'quality' (target) to focus on chemical properties
# We want to cluster wines based on their ingredients, not their rating
X = df.drop(['type', 'quality'], axis=1)

# Scale the data (Critical for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. ELBOW METHOD (Find optimal K)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method (Wine Dataset)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.savefig('wine_elbow.png') # Save for your report
plt.show()

# 4. TRAIN K-MEANS MODEL
# Based on the Elbow, K=3 or K=4 usually works well for this data. Let's use K=3.
k_optimal = 3
kmeans = KMeans(n_clusters=k_optimal, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_scaled)

# 5. VISUALIZATION (Using PCA)
# Since we have 11 columns, we can't plot them all. We use PCA to squash them into 2D.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,7))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y_kmeans, palette='viridis', s=50)
plt.title(f'Wine Clusters (PCA Projection, K={k_optimal})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.savefig('wine_clusters.png') # Save for your report
plt.show()

# 6. EVALUATION
score = silhouette_score(X_scaled, y_kmeans)
print(f"Silhouette Score: {score:.4f}")
```
